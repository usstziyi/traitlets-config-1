### 总结

HelloApp 共拥有 **22 个 traitlets 属性**，其中 **6 个** 标记了 config=True（可通过命令行 -- 参数或配置文件配置）。如果需要查看完整列表，可以在代码中运行：

```python
HelloApp.class_traits()   # 所有 traitlets
HelloApp.class_own_traits()  # 仅 HelloApp 自身定义的
```
